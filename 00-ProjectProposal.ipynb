## Background and Prior Work


Sleep plays a critical role in cognitive functioning, including attention, memory consolidation, and learning—processes that are essential for academic success. College students are particularly vulnerable to irregular sleep patterns due to academic demands, social schedules, and lifestyle habits, making sleep a significant factor in their academic outcomes.

Prior research has consistently demonstrated a strong relationship between sleep and academic performance. S. P. Gilbert and C. C. Weaver (2010) found that poor sleep quality is significantly associated with lower GPA among university students. Similarly, Giuseppe Curcio, Michele Ferrara, and Luigi De Gennaro (2006) conducted a comprehensive review showing that sleep deprivation negatively impacts attention, working memory, and learning capacity.

More recent work by Shannon Hershner and Ronald Chervin (2014) highlights that chronic sleep restriction in college populations leads to decreased alertness and impaired academic functioning. These findings suggest that both sleep quantity and quality are critical contributors to academic success.

While existing studies establish a correlation between sleep and academic performance, fewer studies explore predictive relationships using real-world datasets. This project builds on prior literature by examining whether sleep patterns can be used not only to explain but also to predict academic outcomes over time, while accounting for potential confounding variables.

## Hypothesis


Students who obtain longer and more consistent sleep will achieve higher GPAs and course grades compared to students with shorter or irregular sleep patterns.

Additionally, we hypothesize that improvements in sleep duration and consistency over time will predict measurable improvements in academic performance.

This hypothesis is grounded in prior research showing that sleep enhances memory consolidation, attention, and executive functioning—all of which are critical for learning and academic success.

## Data

An ideal dataset would include longitudinal data (tracked over time) with the following components:

Sleep Variables (Independent Variables):
- Nightly sleep duration (hours)
- Sleep quality (self-reported or device-measured)
- Sleep consistency (variance in sleep schedule)

Academic Performance (Dependent Variables):
- GPA (term-level)
- Course grades or exam scores

Control Variables:
- Study habits (hours studied per week)
- Stress levels
- Caffeine consumption
- Demographics (age, gender, year in school)

Structure:
- One row per student per day or week
- Stored in a structured format (CSV or database)

Real Datasets:
- Sleep Health and Academic Performance (Kaggle)
- Includes sleep duration, quality, stress, and academic metrics
- Suitable for correlation and basic predictive modeling
- National Sleep Research Resource (NSRR)
- Large-scale datasets with detailed physiological sleep data
- Strong for sleep analysis, though academic variables may be limited

Ideal vs. Real Data Considerations:
- Real datasets may lack full longitudinal tracking or complete academic measures
- Some variables (e.g., study habits) may be missing or self-reported
- Despite limitations, these datasets still allow meaningful analysis of relationships between sleep and academic outcomes

Data Collection
- Informed consent: Students must voluntarily agree to share their sleep and academic data and be clearly informed about how the data will be collected, used, and stored.
- Collection bias: Participation may be biased toward certain groups (e.g., more motivated or health-conscious students), which could limit how representative the dataset is.
- Limit PII exposure: Personally identifiable information (such as names, student IDs, or emails) will not be collected, or will be removed during preprocessing to protect participant privacy.
  
Data Storage
- Data security: All data will be stored in secure, access-controlled environments (e.g., password-protected systems or approved cloud storage).
- Right to be forgotten: Participants will have the ability to request that their data be removed from the dataset at any time.
- Data retention plan: Data will only be stored for the duration of the project and will be deleted after completion to minimize long-term risk.
  
Analysis
- Missing perspectives: The dataset may not capture all factors that influence academic performance (e.g., mental health, socioeconomic status), so conclusions will be made cautiously.
- Dataset bias: We will evaluate whether certain demographic groups are underrepresented and discuss how this may affect the validity and generalizability of results.
- Honest representation: Visualizations and statistical summaries will accurately reflect the data without exaggerating trends or implying causation where only correlation exists.
- Privacy in analysis: Results will only be presented in aggregate form, ensuring that no individual student can be identified.
- Auditability: All data processing and analysis steps will be documented and reproducible, allowing others to verify results.
